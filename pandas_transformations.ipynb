## **Importar los datos**

```python
import pandas as pd
import requests

url_hosting = 'https://raw.githubusercontent.com/Near-o7/pandas_transformation/598430d03f40e9277ef1ea5d5d061e70307240a5/datos_hosting.json'
url_inmuebles = 'https://raw.githubusercontent.com/Near-o7/pandas_transformation/refs/heads/main/inmuebles_disponibles.json'

response = requests.get(url_hosting)
hosting_json = response.json()

response = requests.get(url_inmuebles)
inmuebles_json = response.json()
```

```python
datos_hosting = pd.DataFrame(hosting_json)
datos_hosting.head()
```

```python
datos_hosting = pd.json_normalize(datos_hosting['info_inmuebles'])
datos_hosting.head()
```

```python
columnas = list(datos_hosting.columns)
columnas
```

```python
datos_hosting = datos_hosting.explode(column=columnas[3:])
datos_hosting
```

```python
datos_hosting.reset_index(drop=True, inplace=True)
datos_hosting.info()
```

```python
import numpy as np

datos_hosting['max_hospedes'] = datos_hosting['max_hospedes'].astype(np.int64)
datos_hosting.info()
```

```python
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
datos_hosting[col_numericas] = datos_hosting[col_numericas].astype(np.int64)
datos_hosting.info()
```

```python
col_numericas = ['evaluacion_general']
datos_hosting[col_numericas] = datos_hosting[col_numericas].astype(np.float64)
datos_hosting.info()
```

```python
col_numericas = ['cuota_deposito', 'precio', 'cuota_limpieza']
datos_hosting[col_numericas] = datos_hosting[col_numericas].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
datos_hosting.head()
```

```python
col_numericas = ['cuota_deposito', 'precio', 'cuota_limpieza']
datos_hosting[col_numericas] = datos_hosting[col_numericas].astype(np.float64)
datos_hosting.info()
```

### **Datos de Texto**

```python
datos_hosting['descripcion_local'] = datos_hosting['descripcion_local'].str.lower()
datos_hosting.head()
```

```python
datos_hosting['descripcion_local'] = datos_hosting['descripcion_local'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex=True)
datos_hosting['descripcion_local'] = datos_hosting['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)', ' ', regex=True)

datos_hosting.head()
```

```python
datos_hosting['descripcion_local'] = datos_hosting['descripcion_local'].str.split()
datos_hosting['descripcion_local']
```

```python
datos_hosting['comodidades'] = datos_hosting['comodidades'].str.replace(r'\{|}|\"', '', regex=True)
datos_hosting['comodidades'] = datos_hosting['comodidades'].str.split(',')
datos_hosting['comodidades']
```
